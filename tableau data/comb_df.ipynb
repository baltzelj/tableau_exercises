```python
import pandas as pd
```

```python
jan_data = pd.read_csv("JC-202401-citibike-tripdata.csv")
feb_data = pd.read_csv("JC-202402-citibike-tripdata.csv")
mar_data = pd.read_csv("JC-202403-citibike-tripdata.csv")
apr_data = pd.read_csv("JC-202404-citibike-tripdata.csv")
```

```python
jan_df = pd.DataFrame(jan_data)
feb_df = pd.DataFrame(feb_data)
mar_df = pd.DataFrame(mar_data)
apr_df = pd.DataFrame(apr_data)
```

```python
bike_data = [jan_df, feb_df, mar_df, apr_df]

bike_df = pd.concat(bike_data)
```
